# Pregunta 9

Cu√°ntas son las emisiones totales en 'libras' s√≥lo para Costa Rica usando el dataframe 'df_emisions_co2'?



```python
# Coloque el c√≥digo que responda a la pregunta aqui exactamente lo que se pregunta

costa_rica = df_emisions_co2[df_emisions_co2["Country"] == "Costa Rica"]

#Sumar las emisiones en kilotoneladas m√©tricas
total_kt = costa_rica["Value"].sum()

#Convertir a libras
total_libras = total_kt * 2204620

print(float(total_libras))
```

# Pregunta 10

Para s√≥lo 'Costa Rica' y suponiendo que las emisiones se estancan a partir del √∫ltimo a√±o reportado (se mantiene igual). Cu√°ntas son las emisiones en 'Kg' s√≥lo para Costa Rica desde el ultimo a√±o registrado hasta el 2025 usando el dataframe 'df_emisions_co2'?


```python
# Coloque el c√≥digo que responda a la pregunta aqui exactamente lo que se pregunta

costa_rica = df_emisions_co2[df_emisions_co2["Country"] == "Costa Rica"]

#√öltimo a√±o registrado
ultimo_anio = costa_rica["Year"].max()

#Emisiones del √∫ltimo a√±o
emision_ultimo_anio = costa_rica[costa_rica["Year"] == ultimo_anio]["Value"].iloc[0]

#A√±os desde el √∫ltimo registro hasta 2025
anios_a_sumar = 2025 - ultimo_anio

#Convertir a kilogramos (1 kt = 1,000,000 kg)
kg_por_anio = emision_ultimo_anio * 1_000_000

total_kg = kg_por_anio * anios_a_sumar

print(total_kg)
```

# Pregunta 12

Para s√≥lo los pa√≠ses de Centroam√©rica ['El Salvador', 'Honduras', 'Guatemala', 'Costa Rica', 'Nicaragua'] obtener un reporte por a√±o de las emisiones totales en Centroam√©rica usando el dataframe 'df_emisions_co2'?
Se espera un dataframe de 2 columnas: 'Year' y 'total_por_a√±o'

```python
# Coloque el c√≥digo que responda a la pregunta aqui exactamente lo que se pregunta

#Lista de pa√≠ses de Centroam√©rica
paises_centroamerica = ['El Salvador', 'Honduras', 'Guatemala', 'Costa Rica', 'Nicaragua']

centroamerica = df_emisions_co2[df_emisions_co2["Country"].isin(paises_centroamerica)]

reporte = centroamerica.groupby("Year")["Value"].sum().reset_index()

reporte = reporte.rename(columns={"Value": "total_por_a√±o"})

print(reporte)
```

# Pregunta 13

Para s√≥lo los pa√≠ses de Centroam√©rica ['El Salvador', 'Honduras', 'Guatemala', 'Costa Rica', 'Nicaragua'] obtener un reporte por a√±o de las emisiones en promedio en Centroam√©rica usando el dataframe 'df_emisions_co2_per_capita'?
Se espera un dataframe de 2 columnas: 'Year' y 'promedio_por_a√±o'

```python
# Coloque el c√≥digo que responda a la pregunta aqui exactamente lo que se pregunta

#Lista de pa√≠ses de Centroam√©rica
paises_centroamerica = ['El Salvador', 'Honduras', 'Guatemala', 'Costa Rica', 'Nicaragua']

centroamerica_pc = df_emisions_co2_per_capita[df_emisions_co2_per_capita["Country"].isin(paises_centroamerica)]

reporte_pc = centroamerica_pc.groupby("Year")["Value"].mean().reset_index()

reporte_pc = reporte_pc.rename(columns={"Value": "promedio_por_a√±o"})

print(reporte_pc)
```

# Pregunta 14

Cu√°l es el promedio de las emisiones per capita para el a√±o 2005 usando el dataframe 'df_emisions_co2_per_capita'?


```python
# Coloque el c√≥digo que responda a la pregunta aqui exactamente lo que se pregunta

#Filtrar a√±o 2005
ano_2005 = df_emisions_co2_per_capita[df_emisions_co2_per_capita["Year"] == 2005]

#Calcular el promedio
promedio_2005 = ano_2005["Value"].mean()

float(promedio_2005)
```

# Pregunta 15

Usa el dataframe  'df_emisions_co2_per_capita' con las columnas Year y promedio_por_a√±o para crear una gr√°fica de l√≠neas que muestre la evoluci√≥n del promedio de emisiones per c√°pita en Centroam√©rica a lo largo del tiempo; el eje X debe ser el a√±o y el eje Y el promedio de emisiones.

```python
# Coloque el c√≥digo que responda a la pregunta aqui exactamente lo que se pregunta

plt.figure(figsize=(12, 5))

plt.plot(reporte_pc["Year"], reporte_pc["promedio_por_a√±o"], marker="o")

plt.title("Promedio de emisiones per c√°pita en Centroam√©rica a trav√©s del tiempo")

plt.xlabel("A√±o")

plt.ylabel("Promedio de emisiones per c√°pita")

plt.show()
```
